### Init

```python
import matplotlib.pyplot as plt
import time

ENV_NAME = 'Pendulum-v0'
```

### Read data(Pendulum)

```python
fo = open(f"./rewards/d_rewards_{ENV_NAME}.txt", "r")
d_rewards = [] 
episodes = []
for line in fo.readlines():                         
    line = line.strip()
    line = line.split(' ')
    d_rewards.append(float(line[len(line)-1]))
    episodes.append(int(line[0]))

fo = open(f"./rewards/o_rewards_{ENV_NAME}.txt", "r")
o_rewards = [] 
for line in fo.readlines():                         
    line = line.strip()
    line = line.split(' ')
    o_rewards.append(float(line[len(line)-1]))
```

### Draw picture(Pendulum)

```python
plt.ion()
plt.cla()
plt.title('DDPG in Pendulum')
plt.plot(episodes, d_rewards, color='red', label='dueling critic')
plt.plot(episodes, o_rewards, color='blue', label='original critic')
plt.legend()
plt.xlabel('episode steps')
plt.ylabel('normalized state-action value')
plt.show()
plt.pause(0.1)
plt.ioff()
plt.show()
```
